```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('data.csv')
df.head()
```

```python
df.iloc[:,5:].describe()
```

'MOVE','TOTAL','HDB','PRIVATE','COMM','INDU','CYCLPATH','BUS_STOP','ROADINT','ROAD_LIN','ENTROPY','MRTDIST','DISTCEN','RAIN','TEMPO','LABORFREE','WENDFREE'

```python
# sns.set_theme(style="whitegrid")

# Pull out a specific subset of networks

sum_columns = ['MOVE','TOTAL','RAIN']
mean_columns = ['HDB','PRIVATE','COMM','INDU','CYCLPATH','BUS_STOP','ROADINT','ROAD_LIN','ENTROPY','MRTDIST','DISTCEN','TEMPO','LABORFREE','WENDFREE']

# in total 7 days.
df_1 = df.groupby(['CELL_ID','DAYS']).sum().loc[:,sum_columns]
df_2 = df.groupby(['CELL_ID','DAYS']).mean().loc[:,mean_columns]
df_total = pd.concat([df_1,df_2],axis = 1)

# drop wuyi and weekends.
df_workday = df_total[np.logical_and(df_total['LABORFREE'] == 0, df_total['WENDFREE'] == 0)]

```

```python
# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(30, 15))

# Draw a violinplot with a narrower bandwidth than the default
sns.violinplot(data=df_total, palette="Set3", bw=.1, cut=1, linewidth=1)

# Finalize the figure
ax.set(ylim=(0, 30))
sns.despine(left=True, bottom=True)
```

```python
# Compute the correlation matrix and average over networks
corr_df = df_total.corr().groupby('MOVE').mean()
corr_df
```
